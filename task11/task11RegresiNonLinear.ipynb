```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("winequalityN.csv")
df.head(10)
```

```python
# Select attributes and target
X = df['alcohol']
y = df['quality']

# Plot the non-linear regression
plt.scatter(X, y, label='Data Points')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.title('Nonlinear Regression - Wine Quality')
plt.legend()
plt.show()
```
